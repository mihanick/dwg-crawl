```python
#test drawSvg
# https://pypi.org/project/drawSvg/
# !pip install drawsvg
# Works on linux only!
# sudo apt-get install libcairo2
import drawSvg as draw
def plotLine(StartPoint, EndPoint):
    d = draw.Drawing(200,100,origin = 'center', displayInline = False)
    d.append(draw.Lines(StartPoint.X, StartPoint.Y, EndPoint.X, EndPoint.Y, close = False, fill='#eeee00', stroke = 'black'))
    d.setPixelScale(2)
    d.rasterize()
    return(d)

class pnt2:
    def __init__(self, x,y):
        self.X = x
        self.Y = y
start = pnt2(100,200)
end = pnt2(0,0)
plotLine(start,end)
```

```python
ids  = db.files.find({'Valid':True}).distinct('FileId')
print(len(ids))
query = {
    'ClassName':
    {
        '$in':[
            'AcDbLine',
            'AcDbText',
            'AcDbRotatedDimension'
        ]
    },
    'FileId':{
        '$in':ids
    }
}

dd = pd.DataFrame(list(db.objects.find(query)))
dd
```
